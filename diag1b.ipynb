```python
# DIAGNOSTIC DATA ANALYTICS USING PYTHON

from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload dataset
uploaded = files.upload()
df = pd.read_csv('/content/large_sales_data.csv')

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract Month
df['Month'] = df['Date'].dt.to_period('M')

# ---------------------------
# 1 Month-wise Sales & Profit
# ---------------------------
month_summary = df.groupby('Month')[['Sales','Profit']].sum()
print("MONTH-WISE SALES & PROFIT")
print(month_summary)

month_summary.plot(marker='p')
plt.title("Month-wise Sales and Profit")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()

# ---------------------------
# 2 Region-wise Profit
# ---------------------------
region_profit = df.groupby('Region')['Profit'].sum()
print("\nREGION-WISE PROFIT")
print(region_profit)

region_profit.plot(kind='bar')
plt.title("Region-wise Profit")
plt.ylabel("Profit")
plt.show()

# ---------------------------
# 3 Category-wise Profit
# ---------------------------
category_profit = df.groupby('Category')['Profit'].sum()
print("\nCATEGORY-WISE PROFIT")
print(category_profit)

category_profit.plot(kind='bar')
plt.title("Category-wise Profit")
plt.ylabel("Profit")
plt.show()

# ---------------------------
# 4 Correlation Analysis (WHY?)
# ---------------------------
print("\nCORRELATION MATRIX")
print(df[['Sales','Profit','Quantity']].corr())
```

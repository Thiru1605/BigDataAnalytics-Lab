```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/large_sales_data.csv")
```

```python
# Region-wise Total Sales
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(5,5))
plt.bar(region_sales.index, region_sales.values)
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
# Total Profit by Region
region_profit = df.groupby("Region")["Profit"].sum()
plt.figure(figsize=(8,5))
plt.bar(region_profit.index, region_profit.values)
plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Grouping sales & profit by region
region_data = df.groupby("Region")[["Sales", "Profit"]].sum()
plt.figure(figsize=(8,5))
plt.bar(region_data.index, region_data["Sales"], label="Total Sales")
plt.bar(region_data.index, region_data["Profit"], label="Total Profit")
plt.title("Sales vs Profit in Each Region")
plt.xlabel("Region")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Category-wise Total Sales
category_sales = df.groupby("Category")["Sales"].sum()
plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values)
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
# SubCategory-wise Total Sales
subcat_sales = df.groupby("SubCategory")["Sales"].sum()
plt.figure(figsize=(10,5))
plt.bar(subcat_sales.index, subcat_sales.values)
plt.title("Total Sales by SubCategory")
plt.xlabel("SubCategory")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# CustomerType-wise Total Sales
cust_sales = df.groupby("CustomerType")["Sales"].sum()
plt.figure(figsize=(5,5))
plt.bar(cust_sales.index, cust_sales.values)
plt.title("Total Sales by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
pay_sales = df.groupby("PaymentMode")["Sales"].sum()
plt.figure(figsize=(8,5))
plt.bar(pay_sales.index, pay_sales.values)
plt.title("Total Sales by Payment Mode")
plt.xlabel("Payment Mode")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
cust_region = df.groupby(["Region", "CustomerType"]).size().unstack()
cust_region.plot(kind="bar", figsize=(10,5))
plt.title("Customer Type in Each Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.legend(title="Customer Type")
plt.tight_layout()
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month

jan_sales = df[df["Month"] == 1]["Sales"].sum()
jan_profit = df[df["Month"] == 1]["Profit"].sum()

feb_sales = df[df["Month"] == 2]["Sales"].sum()
feb_profit = df[df["Month"] == 2]["Profit"].sum()

print("January Sales :", jan_sales)
print("January Profit:", jan_profit)
print("February Sales :", feb_sales)
print("February Profit:", feb_profit)
```

```python
import numpy as np
months = ["January", "February"]
sales = [jan_sales, feb_sales]
profit = [jan_profit, feb_profit]
x = np.arange(len(months))
width = 0.35
plt.figure(figsize=(8,5))
plt.bar(x - width/2, sales, width, label="Sales")
plt.bar(x + width/2, profit, width, label="Profit")
plt.xticks(x, months)
plt.title("January vs February: Sales and Profit Comparison")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.show()
```
